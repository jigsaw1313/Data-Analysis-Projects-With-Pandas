## Join Methods:

Pandas provides four types of join methods to combine two or more DataFrames based on a common column or index. These methods are:  

**1- Inner Join:** Retains only the rows that have matching values in both DataFrames. In other words, only rows with common values in the specified join column(s) are included in the resulting DataFrame.

**2- Left Join:** Retains all rows from the left DataFrame and only includes matching rows from the right DataFrame based on the specified join column(s). If there are no matches in the right DataFrame, the resulting DataFrame will contain missing values (NaN) for those columns.  

**3- Right Join:** Retains all rows from the right DataFrame and only includes matching rows from the left DataFrame based on the specified join column(s). If there are no matches in the left DataFrame, the resulting DataFrame will contain missing values (NaN) for those columns.  

**4- Outer Join:** Retains all rows from both DataFrames and includes matching rows from both DataFrames based on the specified join column(s). If there are no matches in either DataFrame, the resulting DataFrame will contain missing values (NaN) for those columns.  

You can perform a join operation in Pandas using the merge() method, which allows you to specify the type of join to perform (how parameter), the join column(s) in each DataFrame (left_on and right_on parameters), and other parameters such as whether to retain all rows in the original DataFrames (left_index and right_index parameters).

# Cleaning And Transforming the New `Crew` Column

```python
df['crew'][0]
```

```python
df['crew'] = df['crew'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
df['crew'][0]
```

This conversion is done to transform the string representation of a list into an actual list object.
utilizing the isinstance() function to check if the value in the 'crew' column is a string. If it is, the lambda function applies ast.literal_eval() to convert it into a list. If the value is not a string, it sets it as np.nan.

By performing this operation, you're effectively converting the 'crew' column from a string representation to a list representation, making it easier to work with the data

```python
pd.DataFrame(df['crew'][0])
```

By using df['crew'][0], you're selecting the 'crew' column from the df DataFrame and indexing it with [0] to access the first row of the column.

Next, you're using the pd.DataFrame() function to create a new DataFrame from the first row of the 'crew' column. This will create a DataFrame where each element of the first row is placed in a separate column.

```python
df['crew_size'] = df['crew'].apply(lambda x: len(x))
```

The lambda function lambda x: len(x) is applied to each value in the 'crew' column, where x represents each element of the 'crew' list. The len() function is then used to determine the length of the 'crew' list, which gives you the number of crew members for each movie.

By using the apply() method in conjunction with the lambda function, you're able to iterate over each row in the 'crew' column and calculate the size of the 'crew' list for that particular row.

```python
df['crew_size'].value_counts(dropna = False).head(20)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
        return np.nan
```

#### Explaination On `get_director()` :
 iterating over the elements of the input list x using a for loop. Each element represents a crew member.

Within the loop, you're checking if the 'job' attribute of the crew member is equal to 'Director'. If it is, you're returning the value of the 'name' attribute, which represents the director's name.

If none of the crew members have a 'job' attribute equal to 'Director', you're returning np.nan (NumPy's representation of missing values) to indicate that the director's name is not available.

It's important to note that the return np.nan statement should be outside the for loop, as it should only be executed if none of the crew members have a 'job' attribute equal to 'Director'. In your current code, the return statement is within the loop, which may cause the function to return prematurely after checking only the first crew member.

```python
df['director'] = df['crew'].apply(get_director)
```
