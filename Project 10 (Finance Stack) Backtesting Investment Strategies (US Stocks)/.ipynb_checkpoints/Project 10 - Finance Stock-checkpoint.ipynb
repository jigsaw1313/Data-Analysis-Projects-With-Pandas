# Backtesting a Simple Momentum Strategy

 __Positions__:  
* +1: Investing in DJI (__long__ position) <br>
* -1: Short Selling DJI (__short__ position)<br>
* 0: No position (__neutral__)

 __Strategies__:  
-__Buy and Hold__ (Basic Strategy - passive): Initially Investing into DJI and do nothing (Position: +1 on any given day)

 __Simple Momentum__ (active Strategy to be tested): <br>
__a) Investing (+1)__ into DJI tomorrow if todayÂ´s return was __positive__ <br>
__b) Short selling (-1)__ DJI tomorrow if todayÂ´s return was __negative__

```python
df
```

```python
df["Position"] = np.sign(df["DJI_Return"])
```

```python
df
```

This code will add a new column called Position to the df DataFrame that contains the position of the DJI index for each day. The np.sign() function is used to calculate the sign of the DJI_Return column.

The Position column will have a value of 1 if the DJI index is up on a given day and -1 if the DJI index is down on a given day.

The output of the code will be a new DataFrame with the Position column added.

Here is a more detailed explanation of each parameter in the code:

- `df`: This is the DataFrame that contains the data.
- `np.sign()`: This function calculates the sign of a number.

```python
df["Strategy_Ret"] = df['Position'].shift() * df["DJI_Return"]
df
```

This code will add a new column called Strategy_Ret to the df DataFrame that contains the returns of the strategy for each day. The shift() method is used to shift the Position column one day forward. The * operator is used to multiply the Position column and the DJI_Return column.

The Strategy_Ret column will have a value of the return of the strategy on a given day if the position is long on that day and the negative of the return of the strategy on a given day if the position is short on that day.

The output of the code will be a new DataFrame with the Strategy_Ret column added.

```python
df["Strategy_Ret"].add(1, fill_value = 0).cumprod()
```


This code will add 1 to the Strategy_Ret column and then calculate the cumulative product of the column. The fill_value parameter specifies the value that will be used to fill in missing values. In this case, the fill_value parameter is set to 0.

The output of the code will be a new column called Cumulative_Ret that contains the cumulative returns of the strategy.

Here is a more detailed explanation of each parameter in the code:

- `df`: This is the DataFrame that contains the data.
- `Strategy_Ret`: This is the column that contains the returns of the strategy.
- `add(1, fill_value = 0)`: This function adds 1 to the column and fills in missing values with 0.
- `cumprod()`: This function calculates the cumulative product of the column.

```python
df["Strategy"] = df["Strategy_Ret"].add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
df
```

```python
df["Strategy"] = df["Strategy_Ret"].add(1, fill_value = 0).cumprod()
df
```

```python
df[["DJI_Close", "Strategy"]].plot(figsize = (15,10), fontsize = 15)
plt.title("Simple Momentum Strategy", fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
def summary_ann(returns):
    summary = returns.agg(["mean", "std"]).T
    summary["Return"] = summary["mean"] * 252
    summary["Risk"] = summary['std'] * np.sqrt(252)
    summary.drop(columns = ["mean", "std"], inplace = True)
    return summary
```

```python
summary_ann(df[['DJI_Return', "Strategy_Ret"]])
```

__Strategies__:   

-__Buy and Hold__ (Basic Strategy): Initially Investing into DJI and do nothing (Position: +1 on all days)

-__Simple Contrarian__ (Strategy to be tested): <br>
__a) Short Selling (-1)__ DJI tomorrow if todayÂ´s return was __positive__ <br>
__b) Investing (+1)__ into DJI tomorrow if todayÂ´s return was __negative__

```python
data
```

```python
df = data.loc['2010':, "Close"].to_frame()
df["DJI_Return"] = df['Close'].pct_change()
df.columns = ["DJI_Close", "DJI_Return"]
df.dropna(inplace = True)
df
```

This code create a DataFrame called df that contains the closing prices of the Dow Jones Industrial Average (DJI) from 2010 to the present. The Close column will contain the closing prices, and the DJI_Return column will contain the daily returns. The dropna() function will remove any rows that contain missing values.

The output of the code is a DataFrame that contains the closing prices and daily returns of the DJI from 2010 to the present.

Here is an explanation of each function and parameter:

- loc: This specifies the rows and columns of the DataFrame that should be selected. In this case, the rows are selected from 2010 to the present, and the column is Close.
- to_frame(): This converts the Series into a DataFrame.
- pct_change(): This calculates the daily returns of the DataFrame.
- columns: This specifies the columns of the DataFrame. In this case, the columns are DJI_Close and DJI_Return.
- dropna(): This removes any rows that contain missing values.

```python
df["Position"] = -np.sign(df['DJI_Return'])
df
```

```python
df["Strategy_Ret"] = df["Position"].shift() * df['DJI_Return']
df
```

```python
df["Strategy"] = df["Strategy_Ret"].add(1, fill_value = 0).cumprod() *df.iloc[0, 0]
```

```python
df
```

```python
df[['DJI_Close', "Strategy"]].plot(figsize = (15,10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("Simple Contrarian Strategy", fontsize = 20)
plt.show()
```

```python
summary_ann(df[["DJI_Return", "Strategy_Ret"]])
```

# More Complex Strategies & Backtesting Vs. Fitting

```python
data
```

```python
df = data.loc['2010':, 'Close'].to_frame()
df['DJI_Return'] = df['Close'].pct_change()
df.columns = ["DJI_Close", "DJI_Return"]
df.dropna(inplace = True)
df
```

```python
df['Position'] = np.where(df['DJI_Return'] > 0.01, -1, 1)
df
```

```python
df["Strategy_Ret"] = df['Position'].shift() * df["DJI_Return"]
df
```

```python
df["Strategy"] = df['Strategy_Ret'].add(1, fill_value = 0).cumprod() * df.iloc[0,0]
df
```

```python
df[["DJI_Close", "Strategy"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("Simple Contrarian Strategy")
plt.show()
```

```python
summary_ann(df[["DJI_Close", "Strategy_Ret"]])
```

Even if a Strategy __seems to outperform__ the basic Strategy, the following issues need to be considered/analyzed as well: <br>
- Backtesting vs. Fitting -> Was the Strategy __fitted__ and optimized on __historical data__? __Forward Testing__ is required!<br>
- __Transaction Costs__ -> Changing Positions trigger Cost. Include Costs!<br>

# Simple Moving Average SMA (Part 1)

```python
data
```

```python
df = data.loc['2010':, 'Close'].to_frame()
df['DJI_Return'] = df['Close'].pct_change()
df.columns = ["DJI_Close", "DJI_Return"]
df.dropna(inplace = True)
df
```

```python
df["DJI_Close"].rolling(window = 50).mean()
```

This code calculate the 50-day moving average of the closing prices of the Dow Jones Industrial Average (DJI) in the df DataFrame. The moving average is calculated by averaging the closing prices over a window of 50 days.

The output of the code is a new column in the df DataFrame called DJI_Close_MA50 that contains the 50-day moving average of the closing prices.

- rolling(): This calculates the rolling window statistic. In this case, the rolling window statistic is the mean.
- window: This specifies the size of the rolling window. In this case, the window size is 50 days.
- mean(): This calculates the mean of the DataFrame.


>The moving average can be used to smooth out the data and identify trends.
>The moving average can also be used to filter out noise from the data.

```python
df["SMA50"] = df["DJI_Close"].rolling(window = 50).mean()
```

```python
df[["DJI_Close", "SMA50"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("Simple Moving Average")
plt.show()
```

```python
df["SMA200"] = df['DJI_Close'].rolling(window= 200).mean()
```

```python
df
```

```python
df[['SMA50', "SMA200"]].plot(figsize = (15,10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("SMA50 & SMA200 Crossover Strategy")
plt.show()
```
