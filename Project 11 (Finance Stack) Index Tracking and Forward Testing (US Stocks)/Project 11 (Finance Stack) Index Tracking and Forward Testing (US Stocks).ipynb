### `summary_ann` function to calculate Summary Annualized Risk And Return of each Stock

```python
def summary_ann(returns):
    summary = returns.agg(["mean", "std"]).T
    summary["Return"] = summary["mean"] * 252
    summary["Risk"] = summary["std"] * np.sqrt(252)
    summary.drop(columns = ['mean', 'std'], inplace = True)
    return summary
```

```python
summary = summary_ann(ret)
summary
```

The code snippet defines a function summary_ann(returns) that calculates the annualized summary statistics of a given returns DataFrame.

* summary = returns.agg(["mean", "std"]).T: This calculates the mean and standard deviation of the returns DataFrame using the agg() function. The ["mean", "std"] argument specifies the desired summary statistics. The T attribute transposes the resulting DataFrame so that the statistics are represented as columns.

* summary["Return"] = summary["mean"] * 252: This calculates the annualized return by multiplying the mean of the returns by 252. The assumption here is that there are 252 trading days in a year.

* summary["Risk"] = summary["std"] * np.sqrt(252): This calculates the annualized risk (standard deviation) by multiplying the standard deviation of the returns by the square root of 252.

* summary.drop(columns=['mean', 'std'], inplace=True): This drops the "mean" and "std" columns from the summary DataFrame, as they are no longer needed.

* Finally, the function returns the summary DataFrame, which contains the annualized return and risk statistics.

Note that this function assumes the presence of the NumPy library (np) for the np.sqrt() function. Make sure you import the necessary libraries before using this function.

You can call this function by passing a returns DataFrame as an argument, and it will return a summary DataFrame with annualized return and risk statistics.

```python
def tracking_ann(returns):
    summary = returns.agg(["mean", "std"]).T
    summary["Premium"] = summary["mean"] * 252
    summary["TE"] = summary["std"] * np.sqrt(252) #TE -> Tracking Error
    summary.drop(columns = ["mean", "std"], inplace = True)
    return summary
```

Previous cells defines a function called tracking_ann(returns) that calculates the annualized premium and tracking error of a set of returns.

        summary.drop(columns=["mean", "std"], inplace=True)
        return summary
        
* tracking_ann(returns): This is the function definition, where returns is the input parameter representing a DataFrame or Series containing a set of returns.

The resulting summary statistics are stored in the summary DataFrame, transposed using .T to swap rows and columns.
* summary["Premium"] = summary["mean"] * 252: This line calculates the annualized premium by multiplying the mean returns (summary["mean"]) by the number of trading days in a year, which is assumed to be 252.
* summary["TE"] = summary["std"] * np.sqrt(252): This line calculates the tracking error by multiplying the standard deviation of the returns (summary["std"]) by the square root of the number of trading days in a year (252). The tracking error is a measure of the volatility or dispersion of returns relative to a benchmark.
* summary.drop(columns=["mean", "std"], inplace=True): This line removes the "mean" and "std" columns from the summary DataFrame using the drop() method along the columns axis. The inplace=True parameter ensures that the changes are made in-place, modifying the summary DataFrame directly.

### User Defined Function to Track Errors

```python
def tracking_error_gen(data, tracking_stocks, weights, index, start, end):
    return data.loc[start:end, tracking_stocks].dot(weights).sub(ret.loc[start:end, index]).std() * np.sqrt(252)   
```

___________________________________________________
he code snippet you provided defines a function called tracking_error_gen that calculates the tracking error between a portfolio and a specified index over a given time period.

    def tracking_error_gen(data, tracking_stocks, weights, index, start, end):
    return data.loc[start:end, tracking_stocks].dot(weights).sub(ret.loc[start:end,  index]).std() * np.sqrt(252)
    
* `tracking_error_gen`: This is the function definition. It takes six parameters: data, tracking_stocks, weights, index, start, and end.

* `data:` This parameter represents the data that contains the returns of stocks and the index.

* `tracking_stocks`: This parameter is a list or array-like object that contains the column names of the stocks you want to track.

* `weights`: This parameter represents the weights assigned to each stock in the tracking portfolio. It should be aligned with the tracking_stocks list.

* `index`: This parameter represents the column name of the index you want to track.

* `start and end`: These parameters specify the start and end dates for the time period over which you want to calculate the tracking error.

The function calculates the tracking error by performing the following steps:

`data.loc[start:end, tracking_stocks]` selects the subset of the data DataFrame corresponding to the tracking_stocks for the specified time period.

`.dot(weights)` calculates the dot product between the selected stocks' returns and the weights.

`.sub(ret.loc[start:end, index]) `subtracts the returns of the specified index from the portfolio returns.

`.std(`) calculates the standard deviation of the resulting series.

* np.sqrt(252) scales the standard deviation by the square root of 252 to annualize the tracking error.

The result of the calculation, the tracking error, is returned by the function.

### User Defined Function to Visualize 

```python
def visual(data, tracking_stocks, weights, index, start, end):
    track = data.loc[start:end, tracking_stocks].dot(weights).add(1).cumprod().mul(100)
    track.name = "Portfolio"
    index = data.loc[start:end, index].add(1).cumprod().mul(100)
    plt.figure(figsize = (12,8))
    track.plot()
    index.plot()
    plt.legend(fontsize = 20, loc = 2)
    plt.show()
```
